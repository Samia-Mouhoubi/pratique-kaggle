```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

file_path = './laptop_prices.csv'  
laptop_df = pd.read_csv(file_path)

# Afficher les premières lignes du DataFrame pour vérifier les données
print(laptop_df.head())
```

```python
# 1. Distribution des prix des ordinateurs portables
plt.figure(figsize=(10, 6))
sns.histplot(laptop_df['Price_euros'], kde=True)
plt.title('Distribution des prix des ordinateurs portables')
plt.xlabel('Prix (Euros)')
plt.ylabel('Fréquence')
plt.show()
```

```python
# 2. Relation entre la taille de la RAM et le prix
plt.figure(figsize=(10, 6))
sns.scatterplot(data=laptop_df, x='Ram', y='Price_euros', hue='Company')
plt.title('Prix vs RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Prix (Euros)')
plt.show()
```

```python
# 3. Comparaison des prix par fabricant (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(data=laptop_df, x='Company', y='Price_euros')
plt.title('Gamme de prix par fabricant')
plt.xlabel('Fabricant')
plt.ylabel('Prix (Euros)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Relation entre la taille de l\'écran et le prix
plt.figure(figsize=(10, 6))
sns.scatterplot(data=laptop_df, x='Inches', y='Price_euros', hue='Company')
plt.title('Prix vs Taille de l\'écran')
plt.xlabel('Taille de l\'écran (pouces)')
plt.ylabel('Prix (Euros)')
plt.show()
```
